Q6: A mass of 5 kg is placed in an inclined plane. If 10 N force applied on it and the friction on the inclined plane is 0.4. Find, Sin, Cos and its movement for angle 30, 45 and 60 degrees ?

```python
import math
m = 5
F = 10
g = 9.8
Ff = 0.4
angle1 = 30
angle2 = 45
angle3 = 60

#FOR SIN AND COS

Totangle = angle3+angle2+angle1

print("Total Angle: ",(Totangle),'Degree')
print()
#MOVEMENT

movement= math.sin(Totangle)*m*g
print("Movement  :",(movement),'N')
print()
#FOR COS

angle2 = angle1-angle2-angle3
print("Angle2 is ",(angle2),'Degree')
print()
R = math.cos(angle2)* m * g * Ff
print("The Resultant force and movemnent is :" ,(R),'m/s')




```

Q7: Three resiters of 10, 20 and 30 ohms are connected in series and in parallel if 10 V is given to them find current and voltages across them.

```python
#Resistors in Series
R1=10
R2=20
R3=30
V=10 
Rt=R1+R2+R3
I=V/Rt
V1=I*R1
V2=I*R2
V3=I*R3
print("Total Current across series circuit: ",I,"A")
print("Voltage across R1:",V1,"V")
print("Voltage across R2:",V2,"V")
print("Voltage across R3:",V3,"V")
print()

#Resistors in Parallel
TotResistance=1/(1/R1 +1/R2 +1/R3)
TotCurrent=V/TotResistance
I1=V/R1
I2=V/R2
I3=V/R3

print("Total current accross parallel circuit: ",TotCurrent,"A")
print("Current across R1:",I1,"A")
print("Current across R2:",I2,"A")
print("Current across R3:",I3,"A")


print("Time taken to solve: Approx 10 mins")

```

Q8. Find the point of intercept if line 1 coordinates are (1,5), (7, 9) and line 2 coordinates are (2,4) (6,9).

```python
import math
#FOR LINE 1
x1 = 1
x2 = 7
y1 = 7 
y2 = 9

a1 = (y2-y1)/(x2-x1)
print ("Gradient of line 1 : " , a1)

C2 = y1-a1*x1
print ("y intercept of line 1 is: " , C2)
print ("Equation of Line 1 is: 0.6x - y + 4.3 = 0 ")

# For line 2

x3 = 6
x4 = 2
Y3 = 9
y4 = 4

a2 = (y4-Y3)/(x4-x3)
print ("Gradient of line 2 is: " , a2)

C3 = Y3-a2*x3
print ("y intercept of line 1 is: " , C3)
print ("Equation of Line 2 is: 1.25x - y + 1.5 = 0 ")

xCord = (C3-C2)/(a2-a1)
yCord = (C2*a2-C3*a1)/(a2-a1)

print ("The X coordinate of the point is: " , xCord)
print ("The Y coordinate of the point is : " , yCord)
```
